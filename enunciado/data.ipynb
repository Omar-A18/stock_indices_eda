```python
sp_500["Price_1"] = sp_500["Close"].shift(-1) 
#Creamos una columna la cual tiene el precio del dia anterior y eliminamos una fila de esa columna para posteriormente 
#calcular la diferencia diaria entre precios
sp_500.head()
```

```python
sp_500["Price_Diff"] = sp_500["Price_1"] - sp_500["Close"]
#Añadimos la diferencia entre el precio del dia anterior y el precio del dia actual.
sp_500.head()
```

```python
sp_500["Return"] = sp_500["Price_Diff"]/sp_500["Close"]
sp_500.head()
```

```python
sp_500["Direccion"] = [1 if sp_500.loc[row, "Price_Diff"]> 0 else -1 for row in sp_500.index] 
sp_500.head()
```

```python
sp_500["Average_3d"] = (sp_500["Close"] +sp_500["Close"].shift(1) + sp_500["Close"].shift(2))/3
sp_500.head()
```

```python
sp_500["MA_10"] = sp_500["Close"].rolling(10).mean()
sp_500["MA_50"] = sp_500["Close"].rolling(50).mean()
sp_500.head()
```

```python
sp_500["Shares"] = [1 if sp_500.loc[row, "MA_10"] > sp_500.loc[row, "MA_50"] else 0 for row in sp_500.index]
sp_500.head()
```
